## Step 2 - Window Functions

Let's do some new stuff.

1) An alternative to explicitly creating temporary views is to push data frame in to a parameterised query, we wrap the parameter into {curly brackets} and then list parameters and the variable that they represent in a list after the query.
2) SparkSQL supports the WINDOW clause where we can alias the window used in window functions. The use is trivial here but in a more complex situation or with multiple window functions it makes code much more readable.
3) Cool right? I don't need to nest queries or run in steps, within a SELECT statement I can reference any columns defined earlier in the list.
4) We can do it multiple times in the same statement.

```python
toy_production_windowed = spark.sql("""
SELECT 
    production_date
    ,toys_produced
    ,LAG(toys_produced) OVER DateWindow AS previous_day_production
    , toys_produced - previous_day_production AS production_change 
    , ROUND(100.0*production_change/previous_day_production,2) AS production_change_percentage 
FROM {toy_production_parameter}
WINDOW DateWindow AS (ORDER BY production_date ASC)
"""
,toy_production_parameter=toy_production)

display(toy_production_windowed)
```

## Final Result
The day with the largest percentage increase.


```python
toy_production_answer = spark.sql("""
SELECT *
FROM {toy_production_windowed}
ORDER BY production_change_percentage DESC
LIMIT 1
"""
,toy_production_windowed=toy_production_windowed
)

display(toy_production_answer)
```
