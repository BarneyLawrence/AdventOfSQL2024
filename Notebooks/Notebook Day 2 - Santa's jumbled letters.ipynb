Add each raw table to a data frame.

```python
letters_a = spark.sql("SELECT *, 'Table A' AS TableSource FROM santa_workshop.letters_a")
letters_b = spark.sql("SELECT *, 'Table B' AS TableSource FROM santa_workshop.letters_b")
```

```python
from pyspark.sql.functions import char

letters_combined = letters_combined.withColumn("Character",char(letters_combined.value))
#display(letters_combined.orderBy("TableSource","id"))
```
